```python
def convert(lang: str, input_path):
    
    with codecs.open(input_path, 'r', 'utf-8') as f:
        library = json.load(f)

    nlp_zh = spacy.blank(lang)
    db = DocBin()
    stock_list = []
    lang_list = []
    txt_list = []
    ner_list = []
    check = ''
    check2 = ''
    temp_doc = ''
    
    for document in library:
        if (document['language'] == lang):
          temp1 = document['stock_code']
          temp2 = document['language']

          for txt in document['text']:
              
              temp3 = txt['text']
              
              for ner in ast.literal_eval(txt['ner']):
                  stock_list.append(temp1)
                  lang_list.append(temp2)
                  txt_list.append(re.sub(r'(?<=[^\W\d_])\s+(?=[^\W\d_])', '', temp3))
                  #print(ner)
                  ner_list.append(dict(ner.copy()))
                  
    df1 = pd.DataFrame(list(zip(stock_list, lang_list, txt_list)),
                            columns=['Stock Code', 'Language', 'Paragraph'])
    df1.drop_duplicates(inplace=True)

    df1['Paragraph'] = df1.groupby(['Stock Code','Language'])['Paragraph'].transform(lambda x: ','.join(x))
    df1.drop_duplicates(inplace=True, ignore_index=True)

    return df1
```

```python
df = convert('zh', './data/annual_report_23.json')
df.to_excel('df.xlsx')
df.head()
```

```python
def convert2(cleansed_df, lang: str, output_path):
    
    nlp_zh = spacy.blank(lang)
    db = DocBin()

    for txt in cleansed_df['Paragraph']:
      doc = nlp_zh.make_doc(txt.strip())
      db.add(doc)
     

    db.to_disk(output_path)
    
    return db
```

```python
doc_bin = convert2(df, 'zh', './data/parsed.spacy')
doc_ls = list(doc_bin.get_docs(nlp_zh.vocab))
```

```python
ents_ls = []
for doc in doc_ls:
    document = nlp_zh(doc)
    ents_ls.append([(ent.text, ent.label_) for ent in document.ents])
```

```python
import pickle

with open('./data/ents_ls.pkl', 'wb') as f:
  pickle.dump(ents_ls, f)
```
